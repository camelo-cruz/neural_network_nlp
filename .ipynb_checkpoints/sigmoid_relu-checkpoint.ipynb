```python
import numpy as np
import matplotlib.pyplot as plt
from model.model_utils import relu, relu_prime
```

```python
x = np.arange(-7, 7, 0.1)
y = relu(x)
d_y = relu_prime(y)

plt.subplot(1, 2, 1)
plt.plot(x, y, label='ReLU')
plt.subplot(1, 2, 2)
plt.plot(x, d_y, label='ReLU')     
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def d_sigmoid(z):
    return sigmoid(z) * (1 - sigmoid(z))

y_soft = sigmoid(x)
d_soft = d_sigmoid(x)

plt.subplot(1, 2, 1)
plt.plot(x, y_soft, label='sigmoid')
plt.subplot(1, 2, 2)
plt.plot(x, d_soft, label='sigmoid')
```

# Discussion

Relu is a very simple activation function computationally not expensive as it  is a very simple function to compute. This activation function introduces sparse activation in the network as it sets to 0 all negative values, this is good for training the network. However if most of the outputs are negative, the activations values are going to be set to 0 and the network won't learn anything as the relu derivative for 0 is 0.

Sigmoid is suitable when we want to interpret the output in a probabilistic way because it squishes the values between 0 and 1, so easy to interpret as probabilities. However, the sigmoid activation function can suffer from the gradient vanishing problem as the gradient comes closer and closer to 0 when then activation value is small.
